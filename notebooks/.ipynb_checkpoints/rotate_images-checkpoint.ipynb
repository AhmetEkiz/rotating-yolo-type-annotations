```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
example_image_PATH = '../images/vertical_1.jpg'
image = cv2.imread(example_image_PATH)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# Function to rotate an image and save it to a directory

# def rotate_and_save_image(input_path, output_directory):

def rotate_and_save_image(input_path, output_directory):
    # Load the image
    image = cv2.imread(input_path)

    # Rotate the image by 90 degrees clockwise
    rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

    # # # Construct the output path
    # output_path = os.path.join(output_directory, os.path.basename(input_path))

    # # Save the rotated image to the output directory
    # cv2.imwrite(output_path, rotated_image)
    
    return rotated_image
```

```python
rotated_image = rotate_and_save_image(input_path=example_image_PATH, output_directory='')
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
```
